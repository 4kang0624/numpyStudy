```python
import numpy as np

t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t))
plt.grid()
plt.title('sine graph')
plt.xlabel('sec')
plt.ylabel('sin value')
plt.show()
```

```python
import numpy as np

t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t), lw=5, label='sin') # lw - 굵기
plt.plot(t, np.cos(t), 'g', label='cos')
plt.grid()
plt.legend()
plt.title('sine graph')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```
